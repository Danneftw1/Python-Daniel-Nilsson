```python
import matplotlib.pyplot as plt
import numpy as np

#array med 100 nummer jämlikt fördelat mellan -10 till 10
x = np.linspace(-10, 10, 20)
f = lambda x: np.cos(x)

# equivalent
# def f(x)
#   return np.cos(x)
```

## MATLAB style interface

```python
plt.figure(figsize=(20, 10))
plt.plot(f(x), 'o--')
```

```python
x = np.linspace(-10, 10, 2000)

plt.figure(figsize=(50, 20)) # ändra dimensionerna för lådan
plt.plot(x, f(x))
plt.title("cos(x)")
plt.xlabel("X")
plt.ylabel("Y")
# plt.box(false), tar bort boxen
# h-lines och v-lines för kordinatsystem
```

```python
x, f(x)
```

## OOP aproach

```python
# instantiated fig, ax from figure and axes
fig, ax = plt.figure(figsize = (8,2)), plt.axes()
ax.plot(x, f(x)) # ax object has a plot method
ax.set(xlabel = "x", ylabel="y,", title = "$\cos{x}$", xlim = (-5,5), ylim = (-1.2,1.2))
fig.savefig("../assets/cos(x).png", dpi=200) # sparar bilden
```

```python
# MATLAB style
plt.figure(figsize=(30,4))
plt.subplot(2,1,1) # (rows, columns, panel number)
plt.plot(x, f(x))
plt.plot(x, np.sin(x))
plt.title("cos(x), sin(x)")


plt.subplot(2,1,2)
plt.plot(x, f(x))
plt.title("cos(x)")

plt.suptitle("MATTE BROOOR NAJS")
```

```python
fig, ax = plt.subplot(1,2,3, figsize=(16, 5))

ax[0].plot(x, f(x))
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))

ax[0].set(title = "cos(x) & sin(x)");
fig.suptitle("Big title", fontsize = 25)
```

```python
# MATLAB style
plt.figure(figsize=(16,5))
plt.subplot(1,2,1) # (#rows, #columns, panel number)
plt.plot(x, f(x))
plt.plot(x, np.sin(x))
plt.title("cos(x), sin(x)")

plt.subplot(1,2,2)
plt.plot(x, f(x))
plt.title("cos(x)")

plt.suptitle("Trigonometric functions", fontweight = "bold")

# if we would have 2x2 -> ex 3rd ax: ax[1][0]
```
