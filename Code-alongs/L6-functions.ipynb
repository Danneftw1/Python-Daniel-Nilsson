## Functions

```python
#define a function called cuber, it takes in 1 parameter

def cuber(x):
    return x**3 #värdet som vi skriver in längre ned blir x här

# calls the function cuber using 1 argument
print(f"{cuber(5)=}")
cubes = [cuber (i) for i in range(10)]
cubes
```

```python
import matplotlib.pyplot as plt

plt.plot(cubes)
plt.title("5 upphöjt till 3")
```
