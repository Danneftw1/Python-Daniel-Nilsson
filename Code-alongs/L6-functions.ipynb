## Functions
- Vi använder funktioner så vi kan äteranvända kod och göra ändringar lättare

```python
#define a function called cuber, it takes in 1 parameter

def cuber(x):
    return x**3 #värdet som vi skriver in längre ned blir x här

# calls the function cuber using 1 argument
print(f"{cuber(5)=}")
cubes = [cuber (i) for i in range(10)]
cubes
```

```python
import matplotlib.pyplot as plt

plt.plot(cubes)
plt.title("5 upphöjt till 3")
```

```python
def my_mean(number1, number2, number3):  # 3st parametrar
    print(f"{number1=}, {number2=}, {number3=}")
    return (number1+number2+number3)/3

# my_mean(1,3,6) # 3st argument
print(my_mean(1,3,6))
print(my_mean(0,5,4))

double_mean = my_mean(1,2,3)*2
print(double_mean)
```

## Default values

```python
# keyword arguments
def hello (number_of_greetings, name = "buddy", age = 0):
    if float(age) < 0:
        print("Ange en ålder > 0")
        return
    #om den fastnar på if så fortsätter den inte vidare in i for loopen
    for i in range(number_of_greetings):
        print(f"Hej {name}, du är {age} år")

    # if there is no 'return' -> void function
hello(2)

hello(1, "Bella", -2) #här ändrar vi värdet på name och ålder

# age and name are keyword argument
# number_of_greetings is a positional argument
hello(1, age=34)
```

```python
def word_counter(text = ""):
    word_list = text.split()
    return len (word_list)

quote = "Hello there my friends, ya'll really cool"
word_counter(quote)
```

## Arbitrary arguments - *args (finns också 'kvargs' som är "keyword argument")

```python
def my_mean2(*numbers): # "*" betyder att jag kan lägga in hur många positinal arguments som jag vill
    sum_ = 0
    for number in numbers:
        sum_ += number
    return sum_/len(numbers)

my_mean2(1,2,3,4,5)
```

## Draw lines

```python
import math

def draw_lines(x_list, k=1, m=1, title = ""):
    y_list = [k*x+m for x in x_list]

    plt.plot(x_list, y_list)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(title)

#todo find intersection point and plot it

x = list(range(-10,10))
print(x)
draw_lines(x, title = "Lines")
draw_lines(x, k=-2, m=-1, title = "Lines")
```
