## Random module

````python
import random as rnd


dice = rnd.randint(1,6)
number_sixes = 0
number_rolls = 1000000

for i in range(number_rolls):
    dice = rnd.randint(1,6)
    if dice == 6:
        number_sixes += 1

print(f"{number_sixes}")
print(f"Proportion of sixes {number_sixes/number_rolls}")

## finns random att kolla upp för att kolla random och listor
## random w3schools (bra för att läsa olika metoder och funktioner som vi jobbar med)
````

````python
## b)

import matplotlib.pyplot as plt

#creates an empty list
pasta_per_year = []

# loops 140 years - because 2100-1960
for year in range (140):
    pasta_konsumtion = .791*math.exp(.0526*year) #gör sannolikhetsuträkningen
    # append - adds an element to the list
    pasta_per_year.append(pasta_konsumtion)

plt.plot(range(1960, 2100), pasta_per_year) # själva grafen
plt.xlabel("år") # Alla titlar som gör det tydligare vad modellen är om
plt.ylabel("Pastakonsumtion i kg")
plt.title("Pastakonsumption/kg per person per år från 1960")

print("Supermycket PASTA")
````
