## Super
A function used to call a parent's version of a method.

```python
class Animal: # parent class
    def sound(self):
        print("Animal sound")


class Dog(Animal): #inherits from Animal
    def sound(self):
        super().sound()
        print("Woof!")


dog = Dog()

dog.sound()
```

## Encapsulation
Restricting access to certain details of an object.
Hides information that is not supposed to be accessed from outside of the class

```python
class Cat:
    def __init__(self, name, age) -> None:
        self._name = "Cat"
        self._age = 2
    def sound(self):
        print("Meow!")
        
cat = Cat("Cat", 2)

cat._name
```

```python
import matplotlib.pyplot as plt
import pandas as pd

class AutoPlotter:
    def __init__(self, data):
        self.data = data
        self.plot_type = self.determine_plot_type()

    def determine_plot_type(self):
        if isinstance(self.data, pd.Series):
            return 'bar'
        elif isinstance(self.data, pd.DataFrame):
            if 'time' in self.data.columns:
                return 'line'
            else:
                return 'scatter'
        else:
            return 'unknown'

    def plot(self):
        if self.plot_type == 'bar':
            self.data.plot(kind='bar')
        elif self.plot_type == 'line':
            self.data.plot(x='time', y='value')
        elif self.plot_type == 'scatter':
            self.data.plot(x=self.data.columns[0], y=self.data.columns[1], kind='scatter')
        else:
            print("Unsupported data type for plotting.")

        plt.show()

# Usage
# For pd.Series
data_series = pd.Series([1, 2, 3, 4, 5])
plotter1 = AutoPlotter(data_series)
plotter1.plot()

# For pd.DataFrame (Line plot)
data_df_line = pd.DataFrame({'time': [1, 2, 3], 'value': [1, 4, 9]})
plotter2 = AutoPlotter(data_df_line)
plotter2.plot()

# For pd.DataFrame (Scatter plot)
data_df_scatter = pd.DataFrame({'x': [1, 2, 3], 'y': [1, 4, 9]})
plotter3 = AutoPlotter(data_df_scatter)
plotter3.plot()
```
